````python
import pandas as pd
import numpy as np
pop = 10379295

data = {
    "Kommun" : ["Malmö", "Stockholm", "Uppsala", "Göteborg"],
    "Population" :[ 347949, 975551, 233839, 583065]
}


df = pd.DataFrame(data)
print(df)
print ("\n")
print("\n", df[df["Kommun"].str.contains("Göteborg")], "\n")

sorted = df.sort_values("Population", ascending=False)
print(sorted)
print("\n", sorted.head(3))

sorted["Percent"] = ((sorted["Population"] / pop )*100).round(1)

sorted


````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = "./Data/komtopp50_2020.xlsx"

df = pd.read_excel(path, sheet_name="Totalt", header=6, usecols="A:F")
#df.head()
#df.info()
#df.describe()

df.columns = [
    "Rang 2020",
    "Rang 2019",
    "Kommun",
    "Folkmängd 2020",
    "Folkmängd 2019",
    "Förändring"
]

sorted = df.sort_values("Folkmängd 2020", ascending=False)
#print(sorted.head(10))

smallest = sorted.tail()
largest = sorted.head()
#print("Smallest cities\n", smallest)

total20 = df["Folkmängd 2020"].sum()
total19 = df["Folkmängd 2019"].sum()
print("Total population in 2020: ", total20, "\nTotal population in 2019: ", total19)

lands = pd.concat([largest,smallest])
print(lands)
fig, ax = plt.subplots(figsize=(10,6))

lands.plot(
    kind="bar",
    x = "Kommun",
    y = "Folkmängd 2020",
    color = ["steelblue" if rang > 5 else "orange" for rang in lands["Rang 2020"]],
    ax=ax,
)

ax.set_yscale("log")
bars = ax.containers[0] 
ax.bar_label(bars)
ax.set_title("Largest and smallest cities by population")
ax.set_xlabel("Kommun"), ax.set_ylabel("Folkmängd 2020")
plt.xticks(rotation=45, ha="right")


plt.tight_layout()
plt.show()
````

````python
df_male = pd.read_excel(path,sheet_name="Män",header=6,usecols="A:F")
df_female = pd.read_excel(path,sheet_name="Kvinnor",header=6,usecols="A:F")
#After creating the dataframes, we make a function to remname the rows
def columns(dataframe):

    dataframe.columns = [
        "Rang 2020",
        "Rang 2019",
        "Kommun",
        "Folkmängd 2020",
        "Folkmängd 2019",
        "Förändring"
    ]
    return dataframe

df_male = columns(df_male)
df_female = columns(df_female)

#Then we add the new column
df_male["Kön"]= "Man"
df_female["Kön"]= "Kvinna"

#And finally we merge
cols = ["Kommun","Folkmängd 2020","Folkmängd 2019","Förändring","Kön"]
df_merged = pd.concat([df_male, df_female],ignore_index=True)
df_merged.set_index("Kommun", inplace=True)

#For part c we crfeate a new df from the total
df_original= df[df.columns[2:6]]

df_original.columns= ["Kommun","Total Pop 2020", "Total Pop 2019", "Total förändring"]

#Then we merge it using the index as guide to associate the name of the kommun to the index, then we set kommun back as index
df_final = pd.merge(df_merged[cols[1:5]],df_original, left_index=True, right_on="Kommun").set_index("Kommun")

df_final.sort_values("Total Pop 2020",ascending=False).tail(10)

````

````python
df_final = df_final.sort_values("Total Pop 2020", ascending=False)
df_largest = df_final.head(20)
df_smallest = df_final.tail(20)

df_plot_l = df_largest.pivot_table(index="Kommun", columns="Kön", values="Folkmängd 2020")
df_plot_l = df_plot_l.sort_values(by="Kvinna",ascending= False)
print(df_plot_l)
df_plot_s = df_smallest.pivot_table(index="Kommun", columns="Kön", values="Folkmängd 2020")
df_plot_s = df_plot_s.sort_values(by="Kvinna",ascending= False)

x = range(len(df_plot_l))
width = 0.4

img, axs = plt.subplots(1,2, figsize = (10,6))
bars_f = axs[0].bar([i - width/2 for i in x], df_plot_l["Kvinna"], width=width, label = "Kvinna", color = "Red")
bars_m = axs[0].bar([i + width/2 for i in x], df_plot_l["Man"], width=width, label = "Man", color = "steelblue")



axs[1].bar([i-width/2 for i in x], df_plot_s["Kvinna"], width = width, label = "Kvinna", color = "Red")
axs[1].bar([i+width/2 for i in x], df_plot_s["Man"], width=width, label = "Man", color = "steelblue")

axs[0].set_title("Female and male population in the 10 largest cities of Sweden")
axs[1].set_title("Female and male population in the 10 smallest cities in Sweden")

for ax in axs:
    ax.set_xlabel("Kommun")
    ax.set_ylabel("Folkmängd 2020")
    ax.set_xticks(x)
    ax.legend()
    ax.grid(axis = "y", ls = ":", alpha = 0.6)

axs[0].set_xticklabels(df_plot_l.index, rotation = 45, ha="right")
axs[1].set_xticklabels(df_plot_s.index, rotation = 45, ha="right")

plt.tight_layout()
plt.show()
````
